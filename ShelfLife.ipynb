```python
import cv2
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Load the trained YOLO model
model = YOLO("runs/detect/train/weights/best.pt")  # Change path if needed

# Function to check if an item is rotten
def is_rotten(class_name):
    return "rotten" in class_name.lower()

# Function to predict and warn about rotten items
def detect_rotten_items(image_path):
    # Load image
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Run YOLO prediction
    results = model(image_path)

    # Warning message list
    warnings = []

    # Extract detection details
    for result in results:
        for box in result.boxes:
            # Extract bounding box, class, and confidence
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box
            class_id = int(box.cls[0])  # Class ID
            confidence = float(box.conf[0]) * 100  # Convert confidence to %
            class_name = model.names[class_id]  # Class label

            # Determine text color (Red for rotten, Green for fresh)
            color = (0, 255, 0)  # Green for fresh
            if is_rotten(class_name):
                color = (0, 0, 255)  # Red for rotten
                warnings.append(f"⚠ Remove {class_name} from the shelf!")

            # Draw bounding box
            cv2.rectangle(image_rgb, (x1, y1), (x2, y2), color, 2)
            text = f"{class_name} ({confidence:.1f}%)"
            cv2.putText(image_rgb, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

            # Print detection results
            print(f"Detected: {class_name} ({confidence:.1f}%)")

    # Display warnings if any
    if warnings:
        print("\n🚨 WARNING:")
        for warning in warnings:
            print(warning)

    # Display the image with predictions
    plt.figure(figsize=(8, 8))
    plt.imshow(image_rgb)
    plt.axis("off")
    plt.show()

# Example Usage
image_path = "/content/rotten-apple-with-a-big-hole-on-white-background-JHY6EA.jpg"  # Change path if needed
detect_rotten_items(image_path)
```
